### Basic File Operations

#### 1. Creating and Writing to a File

```python
file_name = 'sample.txt'
content = "Hello, this is a sample file.\nIt contains two lines of text."

with open(file_name, 'w') as f:
    f.write(content)

print(f"File '{file_name}' created and written successfully.")
```

#### 2. Reading from a File

```python
file_name = 'sample.txt'

with open(file_name, 'r') as f:
    read_content = f.read()

print(f"Content of '{file_name}':\n{read_content}")
```

#### 3. Appending to a File

```python
file_name = 'sample.txt'
additional_content = "\nThis is an appended line."

with open(file_name, 'a') as f:
    f.write(additional_content)

print(f"Content appended to '{file_name}'.")

with open(file_name, 'r') as f:
    print(f"New content of '{file_name}':\n{f.read()}")
```

#### 4. Deleting a File

```python
import os

file_name = 'sample.txt'

if os.path.exists(file_name):
    os.remove(file_name)
    print(f"File '{file_name}' deleted successfully.")
else:
    print(f"File '{file_name}' does not exist.")
```

### Basic Data Processing with Pandas

```python
import pandas as pd
import os

# 1. Create a simple CSV file
csv_file_name = 'sales_data.csv'
data = {
    'Product': ['A', 'B', 'A', 'C', 'B', 'A', 'C'],
    'Region': ['East', 'West', 'East', 'North', 'West', 'East', 'North'],
    'Sales': [100, 150, 120, 200, 130, 110, 180]
}
df_raw = pd.DataFrame(data)
df_raw.to_csv(csv_file_name, index=False)

print(f"CSV file '{csv_file_name}' created successfully.")

# 2. Read the CSV file into a pandas DataFrame
df = pd.read_csv(csv_file_name)
print("\nOriginal DataFrame:")
display(df)

# 3. Perform basic data processing
# Filter data: Sales greater than 120
filtered_df = df[df['Sales'] > 120]
print("\nFiltered DataFrame (Sales > 120):")
display(filtered_df)

# Group by Product and calculate total sales
grouped_by_product = df.groupby('Product')['Sales'].sum().reset_index()
print("\nTotal Sales by Product:")
display(grouped_by_product)

# Group by Region and calculate average sales
grouped_by_region = df.groupby('Region')['Sales'].mean().reset_index()
print("\nAverage Sales by Region:")
display(grouped_by_region)

# Clean up the created CSV file
if os.path.exists(csv_file_name):
    os.remove(csv_file_name)
    print(f"\nCSV file '{csv_file_name}' deleted.")
```

#### 2. Fetch Web Page Content



```python
import requests

url = 'http://quotes.toscrape.com/' # A website specifically designed for scraping demonstrations

try:
    response = requests.get(url)
    response.raise_for_status() # Raise an HTTPError for bad responses (4xx or 5xx)
    html_content = response.text
    print(f"Successfully fetched content from {url}")
except requests.exceptions.RequestException as e:
    print(f"Error fetching URL {url}: {e}")
    html_content = None
```

#### 3. Parse HTML and Extract Data



```python
from bs4 import BeautifulSoup

if html_content:
    soup = BeautifulSoup(html_content, 'html.parser')

    # Example 1: Extract all quotes and authors
    quotes = []
    for quote_div in soup.find_all('div', class_='quote'):
        text = quote_div.find('span', class_='text').get_text()
        author = quote_div.find('small', class_='author').get_text()
        quotes.append({'text': text, 'author': author})

    print("\n--- Extracted Quotes ---")
    for q in quotes[:5]: # Displaying first 5 quotes
        print(f"'{q['text']}' - {q['author']}")
    if len(quotes) > 5:
        print(f"... and {len(quotes) - 5} more quotes.")

    # Example 2: Extract all top tags
    tags = []
    for tag_item in soup.find_all('span', class_='tag-item'):
        tag_name = tag_item.find('a', class_='tag').get_text()
        tags.append(tag_name)

    print("\n--- Top Tags ---")
    print(tags)
else:
    print("Cannot parse content because HTML content was not fetched.")
```
